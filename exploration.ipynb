```python
import h5py
from conllu import parse_incr
import pandas as pd

# Replace with your actual file paths
hdf5_path = "../embeddings/BERT_test_embeddings_cleaned.hdf5"
txt_path = "../data/en_ewt-ud-test_cleaned.txt"
conllu_path = "../data/en_ewt-ud-test_cleaned.conllu"

index = 1000

# Print embedding info
with h5py.File(hdf5_path, "r") as f:
    print("Keys in HDF5 file:", len(list(f.keys())))
    # key = str(index)  # Convert index to string
    print(f"Shape of embedding at index {index}:", f[list(f.keys())[index]].shape)

# Print the corresponding sentence
with open(txt_path, "r") as fin:
    sentences = fin.readlines()
    print(f"Sentence at index {index}:\n{sentences[index].strip()}")

with open(conllu_path, "r", encoding="utf-8") as fin:
    sentences = list(parse_incr(fin))
    sentence = sentences[index]
    # Filter tokens in the token list (not pandas)
    filtered_tokens = [token for token in sentence if isinstance(token["id"], int)]
    print(filtered_tokens)
```
